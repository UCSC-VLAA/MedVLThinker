```python
name_list = ["PMC-VQA", "PathVQA", "SLAKE", "VQA_RAD"]

path = "AdaptLLM/biomed-VQA-benchmark"
name = "PathVQA"
split = "test"

dataset = datasets.load_dataset(path, name, split=split)
print(f"Dataset {name} loaded with {len(dataset)} examples.")
print(f"columns: {dataset.column_names}")


df = dataset.to_pandas()
df = df.drop(columns=["image"])
pd.set_option("display.max_rows", 6)
display(df)


closed_answer_df = df[df["answer_type"] == "CLOSED"]
# display(closed_answer_df)
print(f"all possible closed label (# {len(closed_answer_df['label'].unique())}): {closed_answer_df['label'].unique()}")

open_answer_df = df[df["answer_type"] == "OPEN"]
print(f"all possible open label (# {len(open_answer_df['label'].unique())}): {list(open_answer_df['label'].unique())}")


# convert all label field to lower case
closed_answer_df["label"] = closed_answer_df["label"].str.lower()
closed_answer_df = closed_answer_df.sort_values(by=["label"])

# pd.set_option("display.max_rows", 6)
# display(closed_answer_df)

save_path = f"misc/{name}_closed_answer.csv"
closed_answer_df.to_csv(save_path, index=False)
```
